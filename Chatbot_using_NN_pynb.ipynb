```python
import numpy as np
import random
import json

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.preprocessing import LabelEncoder
```

```python
intents = {
    "intents": [
        {
            "tag": "greeting" ,
            "patterns" : [
                "Hi","Hey","Hello"
            ],
            "responses": [
                "Hello!!","'Hey! How can i help you today"
            ]
        },
        {
            "tag": "thanks",
            "patterns": [
                "Thanks","Thank you"
            ],
            "responses": [
                "Youâ€™re welcome ğŸ˜Š", "Anytime!!"
            ]
        },
        {
            "tag": "help",
            "patterns": [
                "I need help","Can you help me"
            ],
            "responses": [
                "I can help you", "I am here to help you"
            ]
        }
    ]
}
```

```python
all_patterns = []
all_tags = []

for intent in intents["intents"]:
  tag = intent["tag"]
  for pattern in intent["patterns"]:
    all_patterns.append(pattern)
    all_tags.append(tag)
print("Total samples:" , len(all_patterns))
print("Example X:", all_patterns[:5])
print("Example y:", all_tags[:5])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(all_patterns)

X_sequences = tokenizer.texts_to_sequences(all_patterns)
max_len = max(len(seq) for seq in X_sequences)

X_padded = pad_sequences(X_sequences, maxlen=max_len, padding='post' )
```

```python
lbl_encoder = LabelEncoder()
y_encoded = lbl_encoder.fit_transform(all_tags)

num_classes = len(set(all_tags))

print("Vocabulary size:",len(tokenizer.word_index)+1)
print("Max dequence length:",max_len)
print("Number of classes:", num_classes)
```

```python
model = keras.Sequential([
    layers.Embedding(input_dim=len(tokenizer.word_index) +1,
                     output_dim=16,
                     input_length= max_len),
    layers.GlobalAveragePooling1D(),
    layers.Dense(16,activation='relu'),
    layers.Dense(num_classes,activation='softmax')
])

```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_padded,y_encoded,epochs=100, verbose=1)
```

```python
def get_response(message):
  seq = tokenizer.texts_to_sequences([message])
  padded = pad_sequences(seq,maxlen=max_len,padding='post')

  pred = model.predict(padded,verbose=0)
  tag_index = np.argmax(pred)

  tag= lbl_encoder.inverse_transform([tag_index])[0]

  for intent in intents["intents"]:
    if intent["tag"] == tag:
      return random.choice(intent["responses"])
```

```python
for msg in ["Hi" , "Thank you", "I need your help in my project"]:
  print("You:",msg)
  print("Bot:", get_response(msg))
  print()
```
